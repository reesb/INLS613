```python
file = open("C:\\Users\\Rees\\Downloads\\AFINN-111.txt", "r")
s = file.read()
s
```

```python
import re
outneg = open("C:\\Users\\Rees\\Downloads\\outneg.txt", "w").close()
outpos = open("C:\\Users\\Rees\\Downloads\\outpos.txt", "w").close()

with open("C:\\Users\\Rees\\Downloads\\outneg.txt", "w") as outneg, open("C:\\Users\\Rees\\Downloads\\outpos.txt", "w") as outpos:

    words = re.findall("(?<=\\n)(.*?)(?=\\t\-?\d)", s)
    words.insert(0, "abandon")
    for word in words:
        if (re.search(word+"\\t-", s)):
            outneg.write(word + "\n")
            print("Added " + word + " to outneg.")
        else:
            outpos.write(word + "\n")
            print("Added " + word + " to outpos.")
```
